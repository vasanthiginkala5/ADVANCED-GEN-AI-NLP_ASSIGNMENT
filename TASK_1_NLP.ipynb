TASK 2

```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove emojis
    text = re.sub(r"[^\w\s]", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # Remove short tokens (≤2) except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
sample_inputs = [
    "I have 2 dogs",                         # Numbers
    "This   is    good",                     # Extra spaces
    "soooo goooood!!!",                     # Repeated characters
    "WOW!!! This is GREAT!!!",              # Uppercase
    "Visit http://example.com now",         # URL
    "Contact me at test@gmail.com",         # Email
    "OK I am at home",                      # Short tokens
    "I am not happy",                       # 'not' exception
]
```

```python
for text in sample_inputs:
    tokens, cleaned = preprocess_text(text)

    print("\n---------------------------")
    print("Original Text:", text)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
```

TASK 3

```python
test_sentences = [
    "Get 100% FREE access now!!!",          # Numbers + symbols + caps
    "I absolutely looooved this product 😍😍",  # Repeated chars + emojis
    "Worst service ever... 0/10",           # Numbers + punctuation
    "Call me at 9876543210",                # Only numbers
    "This is THE best course!!!",           # Mixed case
    "Visit https://openai.com now!",        # URL
    "Contact us at support@gmail.com",      # Email
    "Nooooo this is baaad!!!",              # Repeated characters
    "OK OK OK I got it",                    # Short tokens
    "Win $$$ now!!! Limited offer!!!"       # Special characters
]
```

```python
for text in test_sentences:
    tokens, cleaned = preprocess_text(text)

    print("\n-----------------------------------")
    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
```

```python
import numpy as np

for text in test_sentences:
    tokens, cleaned = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(t) for t in tokens]) if tokens else 0

    print("\n-----------------------------------")
    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
```

```python
from collections import Counter

all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

# Count frequency
counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

The most frequent words like "this" and "now" appear multiple times across different sentences, showing they are common in general text.


The least frequent words appear only once and are usually more specific or context-based terms.

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_sentences)

print("All Tokens:\n", output["tokens"])
print("\nCleaned Sentences:\n", output["clean_sentences"])
```

The pipeline processes a list of raw text inputs.


It applies the preprocessing function to each sentence.


It returns:

A combined list of all cleaned tokens

A list of cleaned sentences

```python
edge_cases = [
    "",             # Empty string
    "😂😂😂😂",       # Only emojis
    "1234567890"    # Only numbers
]

for text in edge_cases:
    tokens, cleaned = preprocess_text(text)

    print("\n---------------------------")
    print("Input:", text)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
```

The preprocessing function correctly handles edge cases without errors.


For an empty string, it safely returns empty output.


For only emojis, all characters are removed during cleaning, resulting in no tokens.


For only numbers, numbers are removed, leaving an empty result.


This ensures the pipeline is robust and suitable for real-world noisy data.
